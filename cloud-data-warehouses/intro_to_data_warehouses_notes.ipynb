**Data Warehouse** is a system (incl. processes, technologies and data representations) that enables us to support analytical processes.

### OLTP vs OLAP

#### OLTP
- operational processes
    - find goods and make orders (for customers)
    - stock and find goods (for inventory staff)
    - pick up and deliver goods (for delivery staff)
- operational databases
    - great for operations
    - no redundancy, high integrity
    - BUT too slow for analytics (too many JOINs)

#### OLAP
- analytical processes
    - assess performance of sales staff (for HR)
    - see effect of different sales channels (for marketing)
    - monitory sales growth (for management)
    
## Dimensional modelling

**Fact tables**
- record business events (e.g. order, phone call, book review)
- columns record events recorded in quantifiable metrics (e.g. quantity of item, duration of call, book rating)

**Dimension tables**
- record context of business events (e.g. who, what, where, why, etc)
- columns contain attributes (e.g. store at which item is bought, customer who made the call)

**Fact or dimension dilemma** - Simple rule - fact is usually **numeric and additive**

Fact examples
- comment on article represents event but we can't easily make a statistic out of its content - not a good fact
- invoice number is numeric but adding it doesn't make sense - not a good fact
- total amount of invoice could be added to calculate total sales - good fact

Example dimensions
- data and time are always dimension
- physical locations and their attributes - good dimension candidates
- human roles (e.g. customers, staff) - good dimension candidates
- goods sold - good dimension candidates

## OLAP Cubes
- once we have a star schema, we can create OLAP cubes
- OLAP cube: aggregation of number of dimensions (e.g. movie, branch, month)

**Roll-up**: sum up sales of each city by country (e.g. US, France, less columns in branch dimension)

**Drill-down**: decompose sales of each city into smaller districts (more columns in branch dimension)

**Slice**: Reducing N dimensions to N-1 dimensions by restricting one dimension to a single value (e.g. month='March')

**Dice**: Same dimensions but computing a sub-cube by restricting, some of the values of the dimensions (e.g. month in ['feb', 'mar'] and movie in ['Avatar', 'Batman'] and branch ='NY')

## DWH storage technology

**MOLAP**
- pre-aggregrate the OLAP cubes and save them on a special purpose non-relational database

**ROLAP**
- compute the OLAP cubes on the fly from existing relational databases where the dimensional model resides
